```python
#before anything else make it sure you have a matplotlib library installed in your local machine in python
#anyway, just in case, on your terminal/command prompt kindly type pip install matplotlib

#import the matplotlib.pyplot library
import matplotlib.pyplot as plt

#let the x as our x-axis that contains the days
x = ["Mon","Tues","Wed","Thurs","Fri"]

#let y as our y-zxis that contains my actual sleeping time
y = [5,6.30,7.30,4.30,4]

#let z as another y-axis containing our supposed sleeping time
z = [2.30,2.15,2,1.30,2]

#let's plot the x and y variables I did earlier. 
#In order to make the actual points being highlighted, let's use scatter to highlight it

plt.plot(x,y)
plt.scatter(x,y)

#let's plot the x and z variables above. 
#In order to make the actual points being highlighted, let's use scatter to highlight it
plt.plot(x,z)
plt.scatter(x,z)

#lets put some legend so that we are not confused on our plotting
plt.legend(["My actual sleep time","My supposed sleep time"])

#labeling the x-axis
plt.xlabel("Dates of June 3 to June 7 2019 ")

#labeling the y-axis as well
plt.ylabel("Sleep time (AM)")

#And lastly, make its title. Make it sure that the headling will really catch the attention 
#of the audience hehe
plt.title("My actual sleeping time is not as good as what my supposed sleeping time is")

#show what you have plotted from the above
plt.show()
```
